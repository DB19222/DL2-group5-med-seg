```python
import numpy as np
import os
import json
from scipy import sparse
from skimage import measure
from scipy.ndimage import affine_transform, rotate
import ast

# Base path configuration
uniseg_path = '/home/roan/Documents/DL2/M3D-Seg/M3D_Seg/'
output_path = '/home/roan/Documents/DL2/M3D-Seg/Transformed/'

# Ensure output directory exists
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Specific dataset code
dataset_code = '0000'
json_path = os.path.join(uniseg_path, dataset_code, dataset_code + '.json')

def apply_transformation(matrix, axis=0, translate=(0, 0), angle=0):
    if axis not in [0, 1, 2]:
        raise ValueError("Axis must be 0 (z-axis), 1 (y-axis), or 2 (x-axis).")

    # Define the axes pair for rotation based on the specified axis
    axes_pairs = [(1, 2), (0, 2), (0, 1)]
    rotation_axes = axes_pairs[axis]

    # Rotate and translate
    matrix_rot = rotate(matrix, angle=angle, axes=rotation_axes, reshape=False, order=1)
    offset = np.zeros(3)
    offset[axis] = translate[0]  # Use translate[0] to move along the main axis
    matrix_transformed = affine_transform(matrix_rot, matrix=np.eye(3), offset=offset, order=1)
    return matrix_transformed


# Load the JSON file
with open(json_path, 'r') as f:
    dataset_dict = json.load(f)

# Process only the first entry
entry = dataset_dict['train'][0]
ct_file_path = os.path.join(uniseg_path, entry['image'])
gt_file_path = os.path.join(uniseg_path, entry['label'])

# Load image and ground truth arrays
img_array = np.load(ct_file_path)[0]
allmatrix_sp = sparse.load_npz(gt_file_path)
gt_shape = ast.literal_eval(gt_file_path.split('.')[-2].split('_')[-1])
gt_array = allmatrix_sp.toarray().reshape(gt_shape)

# Apply transformation
gt_transformed = apply_transformation(gt_array[0], axis=2, translate=(5, 5), angle=5)

# Create 3D mesh
verts, faces, _, _ = measure.marching_cubes(gt_transformed, level=0)

# Save to new folder
mesh_output_dir = os.path.join(output_path, dataset_code)
if not os.path.exists(mesh_output_dir):
    os.makedirs(mesh_output_dir)

mesh_file_path = os.path.join(mesh_output_dir, 'transformed_mesh_first_entry.npz')
np.savez(mesh_file_path, verts=verts, faces=faces)

print(f'Mesh saved for dataset {dataset_code}, first entry. Located at {mesh_file_path}')
```

```python
import matplotlib.pyplot as plt

# Choose a specific slice index to visualize the effect
slice_index = 45

# Visualization of the slices
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gt_array[0, :, :, slice_index], cmap='gray')
plt.title('Original Ground Truth')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gt_transformed[:, :, slice_index], cmap='gray')
plt.title('Transformed Ground Truth')
plt.axis('off')

plt.show()

# Loading mesh data for visualization
original_mesh_data = measure.marching_cubes(gt_array[0], level=0)
transformed_mesh_data = measure.marching_cubes(gt_transformed, level=0)

# Visualization function for 3D mesh
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def plot_3d_mesh(verts, faces, title='3D Mesh'):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    mesh = Poly3DCollection(verts[faces])
    mesh.set_edgecolor('k')
    ax.add_collection3d(mesh)
    ax.set_xlim(0, gt_transformed.shape[0])
    ax.set_ylim(0, gt_transformed.shape[1])
    ax.set_zlim(0, gt_transformed.shape[2])
    ax.set_title(title)
    plt.axis('off')

# Plot original and transformed 3D mesh
plot_3d_mesh(original_mesh_data[0], original_mesh_data[1], title='Original 3D Mesh')
plot_3d_mesh(transformed_mesh_data[0], transformed_mesh_data[1], title='Transformed 3D Mesh')
plt.show()
```
