```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the column names expected in the CSV files
columns = [
    'Organ', '0000', '0002', '0003', '0005', '0006', '0007', '0008', '0009', '0010',
    '0012', '0013', '0014', '0015', '0016', '0017', '0018', '0019', '0020', '0021',
    '0022', '0023', '0024'
]

# Read the data from the CSV files, specifying the column names and skipping initial spaces
df_r2 = pd.read_csv('dice_scores_exp_r2.csv', names=columns, skiprows=1, skipinitialspace=True)
df_r3 = pd.read_csv('dice_scores_exp_r3.csv', names=columns, skiprows=1, skipinitialspace=True)

# Drop any columns that are completely empty
df_r2.dropna(axis=1, how='all', inplace=True)
df_r3.dropna(axis=1, how='all', inplace=True)

# Set the 'Organ' column as the index
df_r2.set_index('Organ', inplace=True)
df_r3.set_index('Organ', inplace=True)

# Calculate the mean Dice scores for each organ, excluding zero values
mean_scores_r2 = df_r2.replace(0, pd.NA).mean(axis=1)
mean_scores_r3 = df_r3.replace(0, pd.NA).mean(axis=1)

# Convert radians to degrees for the title
rotations_degrees = (
    round(0.1 * (180 / 3.14159), 2),
    round(0.1 * (180 / 3.14159), 2),
    round(0.785398163 * (180 / 3.14159), 2)
)

# Plotting the mean Dice scores for each organ
plt.figure(figsize=(14, 8))

# Bar plots for r2 (blue) and r3 (red) with better spacing
bar_width = 0.35
positions_r2 = range(len(mean_scores_r2))
positions_r3 = [p + bar_width for p in positions_r2]

plt.bar(positions_r2, mean_scores_r2, width=bar_width, color='skyblue', label='Baseline (No Rotations)')
plt.bar(positions_r3, mean_scores_r3, width=bar_width, color='salmon', label=f'Rotations: {rotations_degrees[0]}°, {rotations_degrees[1]}°, {rotations_degrees[2]}°')

# Annotate the bars with the mean values for r2
for i, v in enumerate(mean_scores_r2):
    if not pd.isna(v):
        plt.text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom', color='blue')

# Annotate the bars with the mean values for r3
for i, v in enumerate(mean_scores_r3):
    if not pd.isna(v):
        plt.text(i + bar_width, v + 0.01, f'{v:.3f}', ha='center', va='bottom', color='red')

# Customize the plot
plt.xlabel('Organ')
plt.ylabel('Mean Dice Score')
plt.title(f"Mean Dice Scores for Each Organ\nBaseline vs. Rotations: {rotations_degrees[0]}°, {rotations_degrees[1]}°, {rotations_degrees[2]}°")
plt.xticks([p + bar_width / 2 for p in positions_r2], mean_scores_r2.index, rotation=45)
plt.legend(title='Experiments', loc='upper right', ncol=1)
plt.tight_layout()
plt.savefig('mean_dice_scores_for_each_organ_combined.png')
plt.show()
```
